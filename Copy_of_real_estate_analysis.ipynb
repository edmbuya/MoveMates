```python
# 1) Inspect current columns
print("All columns:\n")
for i, c in enumerate(df.columns):
    print(i, repr(c))

# 2) Find mean-price and number-of-dwelling columns by fuzzy match
cols = df.columns.tolist()

mean_price_cols = [c for c in cols if "Mean price" in str(c) or "MeanPrice" in str(c) or "Mean price of residential dwellings" in str(c)]
num_dwell_cols = [c for c in cols if "Number of residential dwellings" in str(c) or "NumDwellings" in str(c) or "Number of residential dwellings" in str(c)]

print("\nDetected mean price columns:", mean_price_cols)
print("Detected number-of-dwelling columns:", num_dwell_cols)

# 3) Build a safe mapping from detected columns to clean names (if detected)
rename_map = {}
state_keys = ["NSW","VIC","QLD","SA","WA","TAS","NT","ACT","AUS"]

# helper to assign state suffix by checking state strings in the original column name
def detect_state_label(colname):
    s = str(colname)
    for st in state_keys:
        if st in s or st.lower() in s.lower() or st.replace(" ","") in s:
            return st
    # fallback: try detect by order if pattern suggests a block
    return None

# Map mean price columns
for c in mean_price_cols:
    st = detect_state_label(c) or ""
    new = f"MeanPrice_{st}" if st else f"MeanPrice_{c}"
    rename_map[c] = new

# Map num dwelling columns
for c in num_dwell_cols:
    st = detect_state_label(c) or ""
    new = f"NumDwellings_{st}" if st else f"NumDwellings_{c}"
    rename_map[c] = new

# Show proposed renames
print("\nProposed renames:")
for k,v in rename_map.items():
    print(repr(k), "->", v)

# Apply renames (only those found)
df = df.rename(columns=rename_map)

# 4) If the automatic detection failed, fall back to manual mapping using column index range:
if not mean_price_cols:
    # fallback indices based on earlier inspection (these may need slight adjustment)
    fallback_mean_indices = list(range(27, 36))  # adjust if necessary
    for idx, st in zip(fallback_mean_indices, state_keys[:len(fallback_mean_indices)]):
        try:
            rename_map[df.columns[idx]] = f"MeanPrice_{st}"
        except Exception:
            pass
    df = df.rename(columns=rename_map)

if not num_dwell_cols:
    fallback_num_indices = list(range(36, 45))  # adjust if necessary
    for idx, st in zip(fallback_num_indices, state_keys[:len(fallback_num_indices)]):
        try:
            rename_map[df.columns[idx]] = f"NumDwellings_{st}"
        except Exception:
            pass
    df = df.rename(columns=rename_map)

print("\nFinal columns head (first 20):")
print(df.columns[:20])

# 5) Now plot mean price with the actual available names (pick states that exist)
import matplotlib.pyplot as plt

available_mean = [c for c in df.columns if c.startswith("MeanPrice_")]
print("\nUsing mean-price columns:", available_mean)

plt.figure(figsize=(12,6))
for c in available_mean:
    if c in df:
        plt.plot(df['Date'], pd.to_numeric(df[c], errors='coerce'), label=c.replace("MeanPrice_",""))
plt.title('Mean Dwelling Price by Region')
plt.xlabel('Date')
plt.ylabel('Mean Price (units in file)')
plt.legend()
plt.grid(True)
plt.show()

# 6) And number of dwellings
available_num = [c for c in df.columns if c.startswith("NumDwellings_")]
print("\nUsing num-dwellings columns:", available_num)

plt.figure(figsize=(12,6))
for c in available_num:
    plt.plot(df['Date'], pd.to_numeric(df[c], errors='coerce'), label=c.replace("NumDwellings_",""))
plt.title('Number of Dwellings by Region')
plt.xlabel('Date')
plt.ylabel('Units (as in file)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# --- Step 1: Calculate YoY Growth Rates ---

# Create new DataFrames for growth rates
price_growth = df.copy()
dwellings_growth = df.copy()

# Identify relevant columns
price_cols = [col for col in df.columns if col.startswith("MeanPrice_")]
num_cols = [col for col in df.columns if col.startswith("NumDwellings_")]

# Compute percentage change
for col in price_cols:
    price_growth[col] = df[col].pct_change() * 100

for col in num_cols:
    dwellings_growth[col] = df[col].pct_change() * 100

# Keep only relevant columns + Date
price_growth = price_growth[["Date"] + price_cols]
dwellings_growth = dwellings_growth[["Date"] + num_cols]

# Drop first NaN row from pct_change
price_growth = price_growth.dropna().reset_index(drop=True)
dwellings_growth = dwellings_growth.dropna().reset_index(drop=True)

# Preview results
print("✅ Mean Price Growth (%):")
print(price_growth.head(3))
print("\n✅ Dwelling Count Growth (%):")
print(dwellings_growth.head(3))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure output directory exists
output_dir = "abs_analysis_outputs"
os.makedirs(output_dir, exist_ok=True)

# --- Use your existing cleaned df ---
# df = your cleaned dataframe from previous steps

# Identify columns
price_cols = [c for c in df.columns if "MeanPrice_" in c]
dwell_cols = [c for c in df.columns if "NumDwellings_" in c]
value_cols = [c for c in df.columns if "Value_" in c]

# 1️⃣ Moving averages (4-quarter)
price_ma = df[["Date"] + price_cols].set_index("Date").rolling(4).mean().reset_index()
dwell_ma = df[["Date"] + dwell_cols].set_index("Date").rolling(4).mean().reset_index()

price_ma.to_csv(f"{output_dir}/meanprice_ma4.csv", index=False)
dwell_ma.to_csv(f"{output_dir}/numdwell_ma4.csv", index=False)

# 2️⃣ Volatility (4-quarter std dev of quarterly % change)
price_growth = df[["Date"]].copy()
for c in price_cols:
    price_growth[c] = df[c].pct_change() * 100
volatility = price_growth.set_index("Date").rolling(4).std().reset_index()
volatility.to_csv(f"{output_dir}/price_volatility.csv", index=False)

# 3️⃣ Value-to-dwelling ratio
value_per_dwelling = df[["Date"]].copy()
for region in [c.replace("Value_", "") for c in value_cols]:
    v_col = f"Value_{region}"
    n_col = f"NumDwellings_{region}"
    if v_col in df.columns and n_col in df.columns:
        value_per_dwelling[region] = df[v_col] / df[n_col]
value_per_dwelling.to_csv(f"{output_dir}/value_per_dwelling.csv", index=False)

# 4️⃣ Rolling correlation between price and dwelling growth (4-quarter window)
rolling_corr = pd.DataFrame({"Date": df["Date"]})
for region in [c.replace("MeanPrice_", "") for c in price_cols]:
    pg = df[f"MeanPrice_{region}"].pct_change()
    dg = df[f"NumDwellings_{region}"].pct_change()
    rolling_corr[region] = pg.rolling(4).corr(dg)
rolling_corr.to_csv(f"{output_dir}/rolling_correlation.csv", index=False)

# 5️⃣ Peak & bottom detection (when growth changes sign)
peaks = {}
for region in [c.replace("MeanPrice_", "") for c in price_cols]:
    growth = df[f"MeanPrice_{region}"].pct_change()
    sign_change = np.sign(growth).diff().fillna(0)
    peaks[region] = {
        "Peaks": df["Date"][sign_change < 0].tolist(),
        "Troughs": df["Date"][sign_change > 0].tolist()
    }

# --- Visualizations ---
sns.set(style="whitegrid")

# Moving average chart for Australia
plt.figure(figsize=(10, 5))
plt.plot(price_ma["Date"], price_ma["MeanPrice_AUS"], label="4Q MA Price (AUS)")
plt.title("Australia: 4-Quarter Rolling Mean of Mean Price")
plt.xlabel("Date"); plt.ylabel("Price ($)")
plt.legend(); plt.show()

# Volatility heatmap (latest year)
latest_vol = volatility.set_index("Date").iloc[-4:].mean().to_frame().T
sns.heatmap(latest_vol[price_cols], cmap="Reds", annot=True, fmt=".2f")
plt.title("Average Price Volatility (Last 4 Quarters)")
plt.show()

# Rolling correlation plot
plt.figure(figsize=(10, 5))
for region in ["NSW", "VIC", "QLD", "WA", "AUS"]:
    plt.plot(rolling_corr["Date"], rolling_corr[region], label=region)
plt.title("4-Quarter Rolling Correlation: Price vs Dwelling Growth")
plt.xlabel("Date"); plt.ylabel("Correlation")
plt.legend(); plt.show()

print("✅ All advanced analytics complete. Outputs saved in 'abs_analysis_outputs/'")
```
